```python
def create_feature(pk_moves):
    if len(pk_moves) == 1:
        pk_moves.append('Nothing')
    return {'first_move':pk_moves[0], 'second_move': pk_moves[1]}
```

```python
features_and_labels = zip(pokemon.abilities, pokemon.type1)
```
